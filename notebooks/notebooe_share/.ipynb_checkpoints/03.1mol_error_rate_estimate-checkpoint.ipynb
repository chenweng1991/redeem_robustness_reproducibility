We use the variants that fail the filtering of max count 2 to estimate the potential background noise level<br/>

It serves as the maximal estimatation of the potential "error rate"

# Load packages and define function

```r
library(redeemR)
library(dplyr)
library(ggplot2)
library(gridExtra)
```

<span style="color: blue;"> Define in-house functions</br>

```r
Estimate_error <-function(redeem,data_bg,edge_trim=5, sample){
bg_var<-subset(data_bg@V.fitered, maxcts==1 & HomoTag=="Hetero")$Variants # define background variants
    ## Count by variant in background, non-edge
    bg_var<-subset(data_bg@V.fitered, maxcts==1 & HomoTag=="Hetero")$Variants
    bg_raw.fragment <- subset(data_bg@raw.fragment.uniqV, variant %in% bg_var)
    bg_raw.fragment_nonedge<- subset(bg_raw.fragment,edge_dist>=edge_trim)
    bg_mut_sumstats<- bg_raw.fragment_nonedge %>% group_by(variant) %>% summarise(pos=unique(pos),countbyvariant=n()) %>% as.data.frame()
    transversion <- subset(bg_raw.fragment,edge_dist>=edge_trim) %>% mutate(changes=add_changes(variant)) %>% mutate(types=add_types(changes))  %>% 
.$types %>% table() %>% prop.table() %>% as.matrix
    
    ## Merge into the total possible mutation table
    info_mut<-subset(data@V.fitered, maxcts>=2 & CellN>=2)$Variants
    bg_nonedge_count_allmut<- subset(ref_all_long, !variant %in% info_mut) %>% 
    mutate(variant=gsub("([0-9]+)([A-Z])>([A-Z])", "\\1_\\2_\\3", variant)) %>% merge(.,bg_mut_sumstats,by="variant",all = T)  
    bg_nonedge_count_allmut <- bg_nonedge_count_allmut %>% mutate(countbyvariant= ifelse(is.na(countbyvariant),0,countbyvariant))

    ## Compute the expected error number
    exp_errorN<-mean(bg_nonedge_count_allmut$countbyvariant)*as.numeric(transversion["transversion",])

    ## Count the number of non-edge 1 mol mutations in observed data
    maxctx2_nonedge_cellN_1mol<-subset(redeem@raw.fragment.uniqV, !variant %in% redeem@HomoVariants & variant %in% info_mut) %>% 
    subset(.,edge_dist>=edge_trim & Freq==1) %>% group_by(variant) %>% summarise(ncell=n(), maxcts=max(Freq))
    
    ## Summarize the number of non-edge 1 mol mutations
    summary(maxctx2_nonedge_cellN_1mol$ncell)
    
    ## Calculate error rate from non-edge 1 mol calls
    frag.non_edge<-subset(redeem@raw.fragment.uniqV, edge_dist>=edge_trim)
    variant_number<-length(unique(subset(frag.non_edge,Freq==1)$variant))
    error_rate<-variant_number*exp_errorN/nrow(subset(frag.non_edge,Freq==1))
    print(error_rate)  
    options(repr.plot.width = 4, repr.plot.height = 4, repr.plot.res = 100)
    p<-ggplot(maxctx2_nonedge_cellN_1mol)+aes(ncell)+geom_histogram(binwidth=0.5,fill="steelblue")+xlim(-1,20)+
    geom_vline(xintercept =exp_errorN,linetype=2 )+theme_cw1()+
    ggtitle(paste0(sample,"-Expected_errors:",round(exp_errorN,4),"\nError rate: <", error_rate))
    summary(maxctx2_nonedge_cellN_1mol$ncell)
    return(p)

   } 
```

> The transversion rate is lower in the non-edge region in this "background" set, suggesting the middle region contains true signals

```r
p1.v1<-Estimate_error(dataset$v1.young1_BMMC$redeemR, dataset_bg$young1_BMMC, edge_trim=5,sample="young1_BMMC")
p2.v1<-Estimate_error(dataset$v1.young1_HPC$redeemR, dataset_bg$young1_HPC, edge_trim=5,sample="young1_HPC")
p3.v1<-Estimate_error(dataset$v1.young1_HSC$redeemR, dataset_bg$young1_HSC, edge_trim=5,sample="young1_HSC")
p4.v1<-Estimate_error(dataset$v1.young2_BMMC$redeemR, dataset_bg$young2_BMMC, edge_trim=5,sample="young1_BMMC")
p5.v1<-Estimate_error(dataset$v1.young2_HPC$redeemR, dataset_bg$young2_HPC, edge_trim=5,sample="young1_HPC")
p6.v1<-Estimate_error(dataset$v1.young2_HSC$redeemR, dataset_bg$young2_HSC, edge_trim=5,sample="young1_HSC")
```
