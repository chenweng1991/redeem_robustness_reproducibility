# Load packages & Functions

```r
library(redeemR)
library(dplyr)
library(stringr)
library(tidyr)
library(purrr)
library(ggsci)
library(cols4all)
library(ggplot2)
library(gridExtra)
draw_mut_type<- function(datatoplot,title){
    datatoplot<- subset(datatoplot, type=="transversion") %>% rbind(., data.frame(label1="exp",label2="exp",type="transversion",sum=0.667))
    datatoplot$label1 <- factor(datatoplot$label1,levels=c("exp","1","2","3+"))
    p<- ggplot(datatoplot)+aes(label1,sum,fill=label2,label=round(sum,2))+geom_bar(stat="identity",position="dodge")+
        geom_text(position = position_dodge(width = 0.9), vjust = -0.5)+theme_classic()+theme_bigger()+ggtitle(title)+labs(y="transversion ratio")
    return(p)
}
get_sum_stat<- function(redeem_raw){
sum_stats<- add_freq_raw(redeem_raw) %>% make_position_df_3.4() %>% mutate(change=sub("^\\d+_", "",variant))
}

## input sum_stat and return the summary of mutation type
do_mut_type<- function(sum_stat,label1="label1",label2="label2"){
    res<-as.data.frame(prop.table(table(sum_stat$change))) %>% mutate(type=ifelse(Var1 %in% c("C_T","G_A","T_C","A_G"),"transition","transversion")) %>%
    .[order(.$type),] %>% mutate(label1=label1,label2=label2)
    return(res)
}
make_position_df_3.4<-function(in_df){
    first <- str_split_fixed(in_df[,"UMI"], "_", 3)[, c(2)] %>% 
        as.numeric()
    last <- str_split_fixed(in_df[,"UMI"], "_", 3)[, c(3)] %>% 
        as.numeric()
    start <- pmin(first, last)
    end <- pmax(first, last)
    df <- data.frame(UMI=in_df[,"UMI"],start = start, end = end, pos = in_df$Pos, 
        variant = in_df$Variants) %>% mutate(length = end - start) %>% 
        mutate(rel_position = (pos - start)/length,Freq=in_df[,"Freq"])
    df$edge_dist<-pmin(abs(df$pos-df$start),abs(df$pos-df$end))
    return(df)
}
## function to from each sumstat
run_mut_type_sumstat<-function(sum_stats,n=4)
{
mut_type_sum<-rbind(do_mut_type(subset(sum_stats,Freq>=3),label1="3+",label2="all"),
      do_mut_type(subset(sum_stats,Freq==2),label1="2",label2="all"),
      do_mut_type(subset(sum_stats,Freq==1),label1="1",label2="all"),
      do_mut_type(subset(sum_stats,Freq>=3 & edge_dist<=n),label1="3+",label2="edge"),
      do_mut_type(subset(sum_stats,Freq==2 & edge_dist<=n),label1="2",label2="edge"),
      do_mut_type(subset(sum_stats,Freq==1 & edge_dist<=n),label1="1",label2="edge"),
      do_mut_type(subset(sum_stats,Freq>=3 & edge_dist>n),label1="3+",label2="non_edge"),
      do_mut_type(subset(sum_stats,Freq==2 & edge_dist>n),label1="2",label2="non_edge"),
      do_mut_type(subset(sum_stats,Freq==1 & edge_dist>n),label1="1",label2="non_edge")
     )
mut_type_sum2plot <- subset(mut_type_sum, label2!="edge") %>% group_by(label1,label2,type) %>% summarise(sum=sum(Freq)) 
return(list(mut_type_sum=mut_type_sum,mut_type_sum2plot=mut_type_sum2plot))
}
theme_bigger <- function(axis_title_size = 14, axis_text_size = 12) {
  theme(
    axis.title = element_text(size = axis_title_size,color="black"),  # Set axis title size
    axis.text = element_text(size = axis_text_size,color="black")     # Set axis text size
  )
}
compute_excess_edge<-function(sumstat,edgecut=4,n=5){
expect_edge<-c()
for(i in 1:n){
    random_df<-make_random(sumstat)
    n<-nrow(random_df)
    expect_edge<-c(expect_edge,nrow(subset(random_df,random_dist_edge<4)))
    }
    excess_percent<-(nrow(subset(random_df,edge_dist<4))- median(expect_edge))/n
    return(excess_percent)
}
make_random<-function(sumstat=all.BMMC){
sumstat$random_pos <- apply(sumstat, 1, randome_pos) %>% unlist
sumstat$random_dist_edge <- pmin(abs(sumstat$random_pos - sumstat$start), abs(sumstat$random_pos - sumstat$end))
return(sumstat)
}    
randome_pos<-function(x){
    sample(x[3]:x[4], 1)
}
```
