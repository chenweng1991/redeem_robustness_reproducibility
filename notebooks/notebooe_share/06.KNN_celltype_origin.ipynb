```r
library(redeemR)
library(ggplot2)
library(dplyr)
library(reshape2)
library(Matrix)
library(amap)
library(gridExtra)
```

```r
## Knn analysis standard
draw_heat_stat<-function(redeem=Young1_BMMC_HSPC_redeem.sensitive_JACnj, filterV=non_info_variants, shuffle=1000, cut=0.05, k=14){
    print(ncol(redeem@Cts.Mtx.bi))
    M<-redeem@Cts.Mtx.bi[,!colnames(redeem@Cts.Mtx.bi) %in% paste("Variants",gsub("_","",filterV),sep="")]
    M<-M[rowSums(M)>0,]
    M<-M[,colSums(M)>0]    
    print(paste("Number of variants to compute distance:",ncol(M)))
    D<-as.matrix(BinaryDist(M,method = "Jaccard"))
    ## Generate celltype.mtx which is a cell vs celltype matrix based on knn
    celltype.mtx<-c()
    Rnames<-row.names(D)
    CellDic<-redeem@CellMeta[,"STD.CellType"]  
    names(CellDic)<-redeem@CellMeta$Cell  
    Cnames<-levels(CellDic)
    for (i in 1:ncol(D)){
        current.cell<-colnames(D)[i]
        name.ordered<-D[,i] %>% .[order(.)] %>% names
        x<-setdiff(name.ordered,current.cell)[1:k]
        celltype.mtx<-rbind(celltype.mtx,as.numeric(table(CellDic[x])))
    }
    row.names(celltype.mtx)<-Rnames
    colnames(celltype.mtx)<-Cnames
    celltype.df<-as.data.frame(celltype.mtx)
    celltype.df$Seed=CellDic[Rnames]
    #Compute aggregated cell type number for each neighbourhood
    NeibourComp_mean<-celltype.df %>% .[complete.cases(.), ] %>% group_by(Seed) %>% dplyr::summarise(MKP=mean(MKP),CMP=mean(CMP),MPP=mean(MPP),HSC=mean(HSC),LMPP=mean(LMPP),Mono=mean(Mono),GMP=mean(GMP),MDP=mean(MDP), cDC=mean(cDC),MEP=mean(MEP),EryP=mean(EryP),ProB=mean(ProB),CLP=mean(CLP),pDC=mean(pDC),B=mean(B),NK=mean(NK),CD8=mean(CD8),CD4=mean(CD4)) 
    cell_order<-c("Mono","MDP","GMP","cDC","EryP","MEP","MKP","HSC","MPP","CMP","pDC","LMPP","CLP","ProB","B","NK","CD8","CD4")
    NeibourComp_mean$Seed<-factor(NeibourComp_mean$Seed,levels=cell_order)
    #Normalization
    m.normed<-NeibourComp_mean[,2:ncol(NeibourComp_mean)] %>% apply(.,2,function(x){x/median(x)}) %>% cbind(NeibourComp_mean[,1],.)
    m.normed$Seed<-paste("S.",m.normed[,1],sep="")
    m.normed<-m.normed[!m.normed$Seed %in% c("S.HSC","S.MPP","S.MKP","S.LMPP","S.CMP"),] %>% select(-HSC,-MPP,-LMPP,-MKP,-CMP)
    m.normed.obs<-m.normed
    hmod<-m.normed.obs %>% tibble::remove_rownames() %>% tibble::column_to_rownames("Seed") %>% Dist(.,method = "correlation") %>% hclust 
    ## Do reshuffle
    celltype.df.reshuffle<-celltype.df
    m.normed.reshuffle.list<-list()
    for(i in 1:shuffle){
        celltype.df.reshuffle$Seed<-sample(celltype.df.reshuffle$Seed)
        NeibourComp_mean<-celltype.df.reshuffle %>% .[complete.cases(.), ] %>% group_by(Seed) %>% dplyr::summarise(MKP=mean(MKP),CMP=mean(CMP),MPP=mean(MPP),HSC=mean(HSC),LMPP=mean(LMPP),Mono=mean(Mono),GMP=mean(GMP),MDP=mean(MDP), cDC=mean(cDC),MEP=mean(MEP),EryP=mean(EryP),ProB=mean(ProB),CLP=mean(CLP),pDC=mean(pDC),B=mean(B),NK=mean(NK),CD8=mean(CD8),CD4=mean(CD4)) 
        NeibourComp_mean$Seed<-factor(NeibourComp_mean$Seed,levels=cell_order)
        m.normed<-NeibourComp_mean[,2:ncol(NeibourComp_mean)] %>% apply(.,2,function(x){x/median(x)}) %>% cbind(NeibourComp_mean[,1],.)
        m.normed$Seed<-paste("S.",m.normed[,1],sep="")
        #Compute m.normed
        m.normed<-m.normed[!m.normed$Seed %in% c("S.HSC","S.MPP","S.MKP","S.LMPP","S.CMP"),] %>% select(-HSC,-MPP,-LMPP,-MKP,-CMP)
        m.normed.reshuffle.list<-c(m.normed.reshuffle.list,list(m.normed))
    }
    #Compute p value
    mtx<-m.normed.obs[,-1]
    pvalue.mtx<-matrix(NA, nrow = nrow(mtx), ncol = ncol(mtx))
    for (i in 1:nrow(mtx)){
        for(j in 1:ncol(mtx)){
            result_vector <- sapply(m.normed.reshuffle.list, function(x) x[i, (j+1)])
            p<-(length(result_vector)-(sum(mtx[i,j]>result_vector)))/length(result_vector)
            pvalue.mtx[i,j]<-p
        }
    }
    colnames(pvalue.mtx)<-colnames(mtx)
    pvalue.mtx<-as.data.frame(pvalue.mtx)
    pvalue.mtx$Seed<-m.normed.obs$Seed
    pvalue.mtx.m<-melt(pvalue.mtx)
    pvalue.mtx.m$ID<-paste(pvalue.mtx.m$Seed,pvalue.mtx.m$variable,sep="_")     
    pvalue.mtx.m$value[pvalue.mtx.m$value>cut]<-NA
    pvalue.mtx.m$value<-as.character(pvalue.mtx.m$value)
    pvalue.mtx.m$value[pvalue.mtx.m$value=="0"]<-"***"
    #
    datatoplot<-melt(m.normed.obs)
    ## prepare datatoplot
    seed.l<-c('S.GMP','S.pDC','S.MEP','S.EryP','S.MDP','S.cDC','S.Mono','S.CD8','S.CD4','S.B','S.NK','S.CLP','S.ProB')
    variable.l<-c('ProB','CLP','NK','B','CD4','CD8','Mono','cDC','MDP','EryP','MEP','pDC','GMP')                                

    datatoplot$Seed<-factor(datatoplot$Seed,levels=seed.l)
    datatoplot$variable<-factor(datatoplot$variable,levels=variable.l)
    datatoplot$ID<-paste(datatoplot$Seed,datatoplot$variable,sep="_")
    datatoplot<-merge(datatoplot,pvalue.mtx.m[,c("ID","value")],by="ID")
    p.main<-ggplot(datatoplot)+aes(variable,Seed,fill=log(value.x),label=value.y)+geom_tile()+scale_fill_gradient2(high="darkred",low="steelblue",mid="white",midpoint=0,na.value = "steelblue",limits=c(-1,1),oob=scales::squish)+
    theme_classic()+theme(axis.text = element_text(size=16,color="black"))+geom_text()
    # plot for reshuffle
    p.reshufles<-list()
    for(i in 1:2){
        datatoplot.reshuffle<-melt(m.normed.reshuffle.list[[sample(1:shuffle,1)]])
        datatoplot.reshuffle$Seed<-factor(datatoplot.reshuffle$Seed,levels=seed.l)
        datatoplot.reshuffle$variable<-factor(datatoplot.reshuffle$variable,levels=variable.l)
        p<-ggplot(datatoplot.reshuffle)+aes(variable,Seed,fill=log(value))+geom_tile()+scale_fill_gradient2(high="darkred",low="steelblue",mid="white",midpoint=0,na.value = "steelblue",limits=c(-1,1),oob=scales::squish)+
        theme_classic()+theme(axis.text = element_text(size=16,color="black"))
        p.reshufles<-c(p.reshufles,list(p))
    }
    p.return<-list(p.main=p.main,p.reshuffles=p.reshufles,hmod=hmod,m.normed.obs=m.normed.obs,m.normed.reshuffle.list=m.normed.reshuffle.list)
    return(p.return)
}
        
                                    ## Knn analysis 1 molecule per cell only
draw_heat_count1only_stat<-function(redeem=Young1_BMMC_HSPC_redeem.sensitive_JACnj,filterV=non_info_variants,shuffle=1000,cut=0.05,k=14){
    M<-redeem@Cts.Mtx[,!colnames(redeem@Cts.Mtx) %in% paste("Variants",gsub("_","",filterV),sep="")]
    M[M>1]<-0 # remove all molecules >1
    M<-M[rowSums(M)>0,]
    M<-M[,colSums(M)>0]
    print(paste("Number of variants to compute distance:",ncol(M)))
    D<-as.matrix(BinaryDist(M,method = "Jaccard"))
    ## Generate celltype.mtx which is a cell vs celltype matrix based on knn
    celltype.mtx<-c()
    Rnames<-row.names(D)
    CellDic<-redeem@CellMeta[,"STD.CellType"]  
    names(CellDic)<-redeem@CellMeta$Cell  
    Cnames<-levels(CellDic)
    for (i in 1:ncol(D)){
        current.cell<-colnames(D)[i]
        name.ordered<-D[,i] %>% .[order(.)] %>% names
        x<-setdiff(name.ordered,current.cell)[1:k]
        celltype.mtx<-rbind(celltype.mtx,as.numeric(table(CellDic[x])))
    }
    row.names(celltype.mtx)<-Rnames
    colnames(celltype.mtx)<-Cnames
    celltype.df<-as.data.frame(celltype.mtx)
    celltype.df$Seed=CellDic[Rnames]
    #Compute aggregated cell type number for each neighbourhood
    NeibourComp_mean<-celltype.df %>% .[complete.cases(.), ] %>% group_by(Seed) %>% dplyr::summarise(MKP=mean(MKP),CMP=mean(CMP),MPP=mean(MPP),HSC=mean(HSC),LMPP=mean(LMPP),Mono=mean(Mono),GMP=mean(GMP),MDP=mean(MDP), cDC=mean(cDC),MEP=mean(MEP),EryP=mean(EryP),ProB=mean(ProB),CLP=mean(CLP),pDC=mean(pDC),B=mean(B),NK=mean(NK),CD8=mean(CD8),CD4=mean(CD4)) 
    cell_order<-c("Mono","MDP","GMP","cDC","EryP","MEP","MKP","HSC","MPP","CMP","pDC","LMPP","CLP","ProB","B","NK","CD8","CD4")
    NeibourComp_mean$Seed<-factor(NeibourComp_mean$Seed,levels=cell_order)
    #Normalization
    m.normed<-NeibourComp_mean[,2:ncol(NeibourComp_mean)] %>% apply(.,2,function(x){x/median(x)}) %>% cbind(NeibourComp_mean[,1],.)
    m.normed$Seed<-paste("S.",m.normed[,1],sep="")
    m.normed<-m.normed[!m.normed$Seed %in% c("S.HSC","S.MPP","S.MKP","S.LMPP","S.CMP"),] %>% select(-HSC,-MPP,-LMPP,-MKP,-CMP)
    m.normed.obs<-m.normed
    hmod<-m.normed.obs %>% tibble::remove_rownames() %>% tibble::column_to_rownames("Seed") %>% Dist(.,method = "correlation") %>% hclust 
    ## Do reshuffle
    celltype.df.reshuffle<-celltype.df
    m.normed.reshuffle.list<-list()
    for(i in 1:shuffle){
        celltype.df.reshuffle$Seed<-sample(celltype.df.reshuffle$Seed)
        NeibourComp_mean<-celltype.df.reshuffle %>% group_by(Seed) %>% dplyr::summarise(MKP=mean(MKP),CMP=mean(CMP),MPP=mean(MPP),HSC=mean(HSC),LMPP=mean(LMPP),Mono=mean(Mono),GMP=mean(GMP),MDP=mean(MDP), cDC=mean(cDC),MEP=mean(MEP),EryP=mean(EryP),ProB=mean(ProB),CLP=mean(CLP),pDC=mean(pDC),B=mean(B),NK=mean(NK),CD8=mean(CD8),CD4=mean(CD4)) 
        NeibourComp_mean$Seed<-factor(NeibourComp_mean$Seed,levels=cell_order)
        options(repr.plot.width=10, repr.plot.height=8,repr.plot.res=100)
        m.normed<-NeibourComp_mean[,2:ncol(NeibourComp_mean)] %>% apply(.,2,function(x){x/median(x)}) %>% cbind(NeibourComp_mean[,1],.)
        m.normed$Seed<-paste("S.",m.normed[,1],sep="")
        #Compute m.normed
        m.normed<-m.normed[!m.normed$Seed %in% c("S.HSC","S.MPP","S.MKP","S.LMPP","S.CMP"),] %>% select(-HSC,-MPP,-LMPP,-MKP,-CMP)
        m.normed.reshuffle.list<-c(m.normed.reshuffle.list,list(m.normed))
    }
    #Compute p value
    mtx<-m.normed.obs[,-1]
    pvalue.mtx<-matrix(NA, nrow = nrow(mtx), ncol = ncol(mtx))
    for (i in 1:nrow(mtx)){
        for(j in 1:ncol(mtx)){
            result_vector <- sapply(m.normed.reshuffle.list, function(x) x[i, (j+1)])
            p<-(length(result_vector)-(sum(mtx[i,j]>result_vector)))/length(result_vector)
            pvalue.mtx[i,j]<-p
        }
    }
    colnames(pvalue.mtx)<-colnames(mtx)
    pvalue.mtx<-as.data.frame(pvalue.mtx)
    pvalue.mtx$Seed<-m.normed.obs$Seed
    pvalue.mtx.m<-melt(pvalue.mtx)
    pvalue.mtx.m$ID<-paste(pvalue.mtx.m$Seed,pvalue.mtx.m$variable,sep="_")     
    pvalue.mtx.m$value[pvalue.mtx.m$value>cut]<-NA
    pvalue.mtx.m$value<-as.character(pvalue.mtx.m$value)
    pvalue.mtx.m$value[pvalue.mtx.m$value=="0"]<-"***"                                
    #
    datatoplot<-melt(m.normed.obs)
    ## prepare datatoplot
    seed.l<-c('S.GMP','S.pDC','S.MEP','S.EryP','S.MDP','S.cDC','S.Mono','S.CD8','S.CD4','S.B','S.NK','S.CLP','S.ProB')
    variable.l<-c('ProB','CLP','NK','B','CD4','CD8','Mono','cDC','MDP','EryP','MEP','pDC','GMP')                                
    datatoplot$Seed<-factor(datatoplot$Seed,levels=seed.l)
    datatoplot$variable<-factor(datatoplot$variable,levels=variable.l)
    datatoplot$ID<-paste(datatoplot$Seed,datatoplot$variable,sep="_")
    datatoplot<-merge(datatoplot,pvalue.mtx.m[,c("ID","value")],by="ID")
    p.main<-ggplot(datatoplot)+aes(variable,Seed,fill=log(value.x),label=value.y)+geom_tile()+scale_fill_gradient2(high="darkred",low="steelblue",mid="white",midpoint=0,na.value = "steelblue",limits=c(-1,1),oob=scales::squish)+
    theme_classic()+theme(axis.text = element_text(size=16,color="black"))+geom_text()
    # plot for reshuffle
    p.reshufles<-list()
    for(i in 1:2){
        datatoplot.reshuffle<-melt(m.normed.reshuffle.list[[sample(1:shuffle,1)]])
        datatoplot.reshuffle$Seed<-factor(datatoplot.reshuffle$Seed,levels=seed.l)
        datatoplot.reshuffle$variable<-factor(datatoplot.reshuffle$variable,levels=variable.l)
        p<-ggplot(datatoplot.reshuffle)+aes(variable,Seed,fill=log(value))+geom_tile()+scale_fill_gradient2(high="darkred",low="steelblue",mid="white",midpoint=0,na.value = "steelblue",limits=c(-1,1),oob=scales::squish)+
        theme_classic()+theme(axis.text = element_text(size=16,color="black"))
        p.reshufles<-c(p.reshufles,list(p))
    }
    p.return<-list(p.main=p.main,p.reshuffles=p.reshufles,hmod=hmod,m.normed.obs=m.normed.obs,m.normed.reshuffle.list=m.normed.reshuffle.list)
    return(p.return)    
}
## A function to test lineage distribution
VariantLinDistri.compute<-function(x,alt="greater"){
    ps<-c()
    Ess<-c()
    for(i in 1:5){
        p<-binom.test(x[i],x[6],Exp[i],alternative = alt)$p.value
        Es<-(x[i]/x[6])/Exp[i]
        ps<-c(ps,p)
        Ess<-c(Ess,Es)
    }
    return(list(ps,Ess)) 
}
```

# Redeem filter 1

### Load data & prepare variants summary

```r
Young1_BMMC_HSPC_redeem.sensitive_JACnj <- readRDS("/lab/solexa_weissman/cweng/Projects/Collaborator/Caleb/github_redeem_plus/data/Young1_BMMC_HSPC_Redeem_filter1.rds")
```

```r
## Prepare variants summary
LV <- Young1_BMMC_HSPC_redeem.sensitive_JACnj@GTsummary.filtered %>% 
        merge(.,Young1_BMMC_HSPC_redeem.sensitive_JACnj@CellMeta[,c("Cell","Lineage")]) 
## Make LV.summarise
LV.summarise <- LV %>% group_by(Variants) %>% dplyr::count(Lineage) %>% dcast(Variants~Lineage)
row.names(LV.summarise) <- LV.summarise$Variants
LV.summarise <- LV.summarise[,-1]
LV.summarise[is.na(LV.summarise)] <- 0
LV.summarise$sum <- rowSums(LV.summarise)
## Calculate expected lineage proportion
Exp <- colSums(LV.summarise[,-6])/sum(colSums(LV.summarise[,-6]))
## Compute metrics to distringuish linegae-informative and non-informative variants
LV.summarise.raw <- apply(LV.summarise,1,VariantLinDistri.compute)
LV.summarise.enrich.pvalues <- lapply(LV.summarise.raw,function(x){x[[1]]}) %>% do.call(rbind,.) %>% as.data.frame
LV.summarise.enrich.effectSizes <- lapply(LV.summarise.raw,function(x){x[[2]]}) %>% do.call(rbind,.)  %>% as.data.frame
## Identify the lineage non-informative variants (remove in knn analysis)
non_info_variants <- LV.summarise.enrich.effectSizes[which((LV.summarise.enrich.effectSizes$Mye<=1.5 | LV.summarise.enrich.pvalues$Mye>=0.05) &
                                      (LV.summarise.enrich.effectSizes$Lym<=1.5 | LV.summarise.enrich.pvalues$Lym>=0.05) &
                                      (LV.summarise.enrich.effectSizes$ME<=1.5  | LV.summarise.enrich.pvalues$ME>=0.05)  &
                                      (LV.summarise.enrich.effectSizes$MKP<=1.5  | LV.summarise.enrich.pvalues$MKP>=0.05)),] %>% row.names 
```

### Run cell type knn analysis

```r
p.all.filter1 <- draw_heat_stat(Young1_BMMC_HSPC_redeem.sensitive_JACnj, non_info_variants)
p.all_count1only.filter1<-draw_heat_count1only_stat(redeem=Young1_BMMC_HSPC_redeem.sensitive_JACnj, filterV=non_info_variants, shuffle=1000, cut=0.05, k=14)
```

```r
options(repr.plot.width=20, repr.plot.height=8,repr.plot.res=150)
grid.arrange(p.all.filter1[[1]]+ggtitle("all"),p.all_count1only.filter1[[1]]+ggtitle("n==1"),nrow=1)
```
