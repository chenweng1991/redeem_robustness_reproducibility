```r
make_position_df_2 <- function(in_df){
  first<-str_split_fixed(in_df[["V1"]], "_", 3)[,c(2)] %>% as.numeric()
  last<-str_split_fixed(in_df[["V1"]], "_", 3)[,c(3)] %>% as.numeric()
  start<-pmin(first,last)
  end<-pmax(first,last)
  df<-data.frame(
    UMI= in_df$V1,
    Cell= in_df$V2,
    start = start,
    end = end,
    pos = in_df$V3, variant = in_df$V4
  ) %>% mutate(dist_edge=pmin(abs(pos-start),abs(pos-end)),length = end - start) %>% 
    mutate(rel_position=(pos-start)/length)
    return(df)
}
Make_MitoTracing_VS <- function (RawGenotypes, depth, limitV = NA, Homo = "", WhitelistV = MKP_WhiteList, 
    Ncell = 2) 
{
    GiveName <- c("UMI", "Cell", "Pos", "Variants", "Call", "Ref", 
        "FamSize", "GT_Cts", "CSS", "DB_Cts", "SG_Cts", "Plus", 
        "Minus", "Depth")
    colnames(RawGenotypes) <- GiveName
    RawGenotypes <- RawGenotypes %>% filter((DB_Cts > 0 & FamSize >= 
        1 & CSS > 0.75) | (DB_Cts == 0 & FamSize >= 2 & CSS > 
        0.75))
    
    VariantsGTSummary <- list(Sensitive = GTSummary(RawGenotypes) %>% 
        mutate(hetero = Freq/depth))
    Variants.feature.lst <- Vfilter_X_Max_Count(InputSummary = VariantsGTSummary, 
        depth = depth, Min_Cells = 2, QualifyCellCut = 10, WhitelistV = WhitelistV)
    Variants.feature.lst$Sensitive <- Variants.feature.lst$Sensitive %>% 
        subset(., !Variants %in% Homo)
    mitoTracing.Sensitive <- Create_mitoTracing(GTsummary_list = list(VariantsGTSummary), 
        depth_list = list(depth), feature.list_list = list(Variants.feature.lst), 
        meta_list = list(depth$Total[[2]] %>% rename(., ATACName = V1)), 
        thr = "Sensitive", labels = c("Test"), qualifiedCellCut = 10, 
        OnlyHetero = T, VAFcut = 1, Cellcut = Ncell, maxctscut = 1)
    require(dplyr)
    require(Matrix.utils)
    Cts.Mtx <- dMcast(mitoTracing.Sensitive@GTsummary.filtered, 
        Cell ~ Variants, value.var = "Freq")
    colnames(Cts.Mtx) <- strsplit(as.character(colnames(Cts.Mtx)), 
        "_") %>% sapply(., function(x) {
        paste(x[1], x[2], x[3], sep = "")
    })
    Cts.Mtx.bi <- Cts.Mtx
    Cts.Mtx.bi[Cts.Mtx.bi >= 1] <- 1
    Cts.Mtx.bi <- Cts.Mtx.bi[which(rowSums(Cts.Mtx.bi)>0), which(colSums(Cts.Mtx.bi)>1)]
    print(dim(Cts.Mtx.bi))
    if (length(limitV) > 1) {
        Cts.Mtx <- Cts.Mtx[, intersect(colnames(Cts.Mtx), limitV)]
        Cts.Mtx.bi <- Cts.Mtx.bi[, intersect(colnames(Cts.Mtx), 
            limitV)]
    }
    mitoTracing.Sensitive@Cts.Mtx.bi <- Cts.Mtx.bi
    mitoTracing.Sensitive@Cts.Mtx <- Cts.Mtx
    data(CellPCT)
    V.weight <- data.frame(weight = 1 - CellPCT$muRate)
    V.weight$Variants <- paste("Variants", gsub("_", "", CellPCT$Variant), 
        sep = "")
    weight <- data.frame(Variants = colnames(mitoTracing.Sensitive@Cts.Mtx.bi)) %>% 
        merge(., V.weight, by = "Variants", all.x = T, sort = F) %>% 
        .$weight
    weight[is.na(weight)] <- 1
    d.w_jaccard <- quick_w_jaccard(mitoTracing.Sensitive@Cts.Mtx.bi, 
        w = weight)
    phylo <- nj(d.w_jaccard)
    return(list(Cts.Mtx=Cts.Mtx,Cts.Mtx.bi = Cts.Mtx.bi, d.w_jaccard = d.w_jaccard, 
        phylo = phylo, mitoTracing.Sensitive = mitoTracing.Sensitive))
}
Make_MitoTracing_VS_count1exclude <- function (RawGenotypes, depth, limitV = NA, Homo = "", WhitelistV = MKP_WhiteList, Ncell = 2) 
{
    GiveName <- c("UMI", "Cell", "Pos", "Variants", "Call", "Ref", 
        "FamSize", "GT_Cts", "CSS", "DB_Cts", "SG_Cts", "Plus", 
        "Minus", "Depth")
    colnames(RawGenotypes) <- GiveName
    RawGenotypes <- RawGenotypes %>% filter((DB_Cts > 0 & FamSize >= 
        1 & CSS > 0.75) | (DB_Cts == 0 & FamSize >= 2 & CSS > 
        0.75))
    VariantsGTSummary <- list(Sensitive = GTSummary(RawGenotypes) %>% 
        mutate(hetero = Freq/depth))
    Variants.feature.lst <- Vfilter_X_Max_Count(InputSummary = VariantsGTSummary, 
        depth = depth, Min_Cells = 2, QualifyCellCut = 10, WhitelistV = WhitelistV)
    Variants.feature.lst$Sensitive <- Variants.feature.lst$Sensitive %>% 
        subset(., !Variants %in% Homo)
    mitoTracing.Sensitive <- Create_mitoTracing(GTsummary_list = list(VariantsGTSummary), 
        depth_list = list(depth), feature.list_list = list(Variants.feature.lst), 
        meta_list = list(depth$Total[[2]] %>% rename(., ATACName = V1)), 
        thr = "Sensitive", labels = c("Test"), qualifiedCellCut = 10, 
        OnlyHetero = T, VAFcut = 1, Cellcut = Ncell, maxctscut = 1)
    require(dplyr)
    require(Matrix.utils)
    Cts.Mtx <- dMcast(mitoTracing.Sensitive@GTsummary.filtered, 
        Cell ~ Variants, value.var = "Freq")
    colnames(Cts.Mtx) <- strsplit(as.character(colnames(Cts.Mtx)), 
        "_") %>% sapply(., function(x) {
        paste(x[1], x[2], x[3], sep = "")
    })
    # Cts.Mtx<-Cts.Mtx[,apply(Cts.Mtx,2,max)>=2]
    Cts.Mtx.bi <- Cts.Mtx
    Cts.Mtx.bi[Cts.Mtx.bi == 1] <- 0
    Cts.Mtx.bi[Cts.Mtx.bi > 1] <- 1
    Cts.Mtx.bi <- Cts.Mtx.bi[which(rowSums(Cts.Mtx.bi)>0), which(colSums(Cts.Mtx.bi)>1)]
    print(dim(Cts.Mtx.bi))
    if (length(limitV) > 1) {
        Cts.Mtx <- Cts.Mtx[, intersect(colnames(Cts.Mtx), limitV)]
        Cts.Mtx.bi <- Cts.Mtx.bi[, intersect(colnames(Cts.Mtx), 
            limitV)]
    }
    mitoTracing.Sensitive@Cts.Mtx.bi <- Cts.Mtx.bi
    mitoTracing.Sensitive@Cts.Mtx <- Cts.Mtx
    data(CellPCT)
    V.weight <- data.frame(weight = 1 - CellPCT$muRate)
    V.weight$Variants <- paste("Variants", gsub("_", "", CellPCT$Variant), 
        sep = "")
    weight <- data.frame(Variants = colnames(mitoTracing.Sensitive@Cts.Mtx.bi)) %>% 
        merge(., V.weight, by = "Variants", all.x = T, sort = F) %>% 
        .$weight
    weight[is.na(weight)] <- 1
    d.w_jaccard <- quick_w_jaccard(mitoTracing.Sensitive@Cts.Mtx.bi, 
        w = weight)
    d.w_jaccard[which(is.infinite(d.w_jaccard))] <- 1
    d.w_jaccard[which(is.nan(d.w_jaccard))] <- 1
    phylo <- nj(d.w_jaccard)
    return(list(Cts.Mtx=Cts.Mtx,Cts.Mtx.bi = Cts.Mtx.bi, d.w_jaccard = d.w_jaccard, 
        phylo = phylo, mitoTracing.Sensitive = mitoTracing.Sensitive))
}
reshuffle_mito_master<-function(ob){
    Mtx<-apply(ob[["Cts.Mtx.bi"]],2,function(col){sample(col)}) %>% as(.,"sparseMatrix")
    row.names(Mtx)<-row.names(ob[["Cts.Mtx.bi"]])
    colnames(Mtx)<-colnames(ob[["Cts.Mtx.bi"]])
    V.weight <- data.frame(weight = 1 - CellPCT$muRate)
    V.weight$Variants <- paste("Variants", gsub("_", "", CellPCT$Variant), sep = "")
    weight <- data.frame(Variants = colnames(Mtx)) %>% merge(., V.weight, by = "Variants", all.x = T, sort = F) %>% .$weight
    weight[is.na(weight)] <- 1
    d.w_jaccard <- quick_w_jaccard(Mtx, w = weight)
    ob[["Cts.Mtx.bi"]]<-Mtx
    ob[["d.w_jaccard"]]<-d.w_jaccard
    return(ob)
} 

make_position_df_3 <- function(in_df){
  first<-str_split_fixed(in_df[["V1"]], "_", 3)[,c(2)] %>% as.numeric()
  last<-str_split_fixed(in_df[["V1"]], "_", 3)[,c(3)] %>% as.numeric()
  dist_edge=pmin(abs(in_df$V3-first),  abs(in_df$V3-last))
  return(dist_edge)
}
RunMito2Crisp_v2<-function (Whm, Jac, Cts.Mtx.bi, k, reshuffleN = 1000) 
{
    Jac.d <- as.matrix(Jac)
    Whm.d <- as.matrix(Whm)
    IntersectCell <- intersect(row.names(Jac.d), row.names(Whm.d))
    Jac.d <- Jac.d[IntersectCell, IntersectCell]
    Whm.d <- Whm.d[IntersectCell, IntersectCell]
    Jac.nn <- MakeNN_self_exclude(Jac.d, k.param = k)
    Crisp_Jac.rank <- apply(Whm.d, 1, rank) %>% t
    n <- nrow(Crisp_Jac.rank)
    pvalues <- c()
    EffectSizes <- c()
    for (i in 1:nrow(Crisp_Jac.rank)) {
        Reshuffle <- c()
        for (j in 1:reshuffleN) {
            Reshuffle <- c(Reshuffle, mean(sample(Crisp_Jac.rank[i, 
                ], (k - 1))))
        }
        Rank <- mean(Crisp_Jac.rank[i, Jac.nn$idx[i, 1:k]])
        p <- length(which(Rank > Reshuffle))/reshuffleN
        Rankpct <- (mean(Reshuffle) - Rank)/n
        pvalues <- c(pvalues, p)
        EffectSizes <- c(EffectSizes, Rankpct)
    }
    Mito2Crisp.df <- data.frame(Cell = row.names(Crisp_Jac.rank), 
        pvalues = pvalues, EffectSizes = EffectSizes)
    Mito2Crisp.df$pvalues[Mito2Crisp.df$pvalues == 0] <- 1/reshuffleN
    row.names(Mito2Crisp.df) <- Mito2Crisp.df$Cell
    Mito2Crisp.df <- Tomerge_v2(Mito2Crisp.df, data.frame(VN = rowSums(Cts.Mtx.bi)))
    return(Mito2Crisp.df)
}
MakeNN_self_exclude<-function (d, k.param = 15) 
{
    d <- as.matrix(d)
    n.cells <- dim(d)[1]
    object <- d
    
    knn.mat <- matrix(data = 0, ncol = k.param, nrow = n.cells)
    knd.mat <- knn.mat
    for (i in 1:n.cells) {
        knn.mat[i, ] <- setdiff(order(object[i,]),i)[1:k.param]
        knd.mat[i, ] <- object[i, knn.mat[i, ]]
    }
    nn.idx <- knn.mat[, 1:k.param]
    nn.dist <- knd.mat[, 1:k.param]
    nn <- list(idx = nn.idx, dist = nn.dist)
    return(nn)
}
Draw_Mito2Crispr2<-function (phylo, Mito2Crispr.df) 
{
    library(viridis)
    library(dplyr)
    p <- ggtree(as.treedata(phylo), branch.length = "none") + 
        geom_fruit(data = Mito2Crispr.df, geom = geom_bar, mapping = aes(y = Cell, 
            x = -EffectSizes, fill = -log10(pvalues)), pwidth = 0.5, 
            width = 1, offset = 0.4, orientation = "y", stat = "identity", 
            axis.params = list(axis = "x", text.size = 2, hjust = 1, 
                vjust = 0.5, nbreak = 3)) + scale_fill_gradient(high = "red", 
        low = "steelblue")
    return(p)
}
theme_bigger <- function(axis_title_size = 14, axis_text_size = 12) {
  theme(
    axis.title = element_text(size = axis_title_size,color="black"),  # Set axis title size
    axis.text = element_text(size = axis_text_size,color="black")     # Set axis text size
  )
}
```

## Load data

```r
# load data
Samples.name <- c("CD24_1", "Ctrl_1")
whm.list <- readRDS("./data/whm.list.rds")
RawGenotype.filter1.list <- readRDS("./data/RawGenotype.list.readeem_filter1.rds")
RawGenotype.filter2.list <- readRDS("./data/RawGenotype.list.readeem_filter2.rds")
MKP_WhiteList <- readRDS("./data/MKP_WhiteList.rds")
MKP.depth <- readRDS("./data/MKP.depth.rds")
names(RawGenotype.filter1.list) <- paste(names(RawGenotype.filter1.list), "filter1",sep='.')
names(RawGenotype.filter2.list) <- paste(names(RawGenotype.filter2.list), "filter2",sep='.')
RawGenotype.list <- c(RawGenotype.filter1.list, RawGenotype.filter2.list)
whm.list <- c(whm.list, whm.list)
names(whm.list) <- names(RawGenotype.list)
```

## Compare Redeem mito mutations & CRISPR 

```r
# 
mito_master.list <- list()
AOC_lis <- list()
phylo_lis <- list()
for(i in 1:length(RawGenotype.list)){
    # All
    cur_name <- names(RawGenotype.list)[i]
    mito_master <- Make_MitoTracing_VS(RawGenotype.list[[i]][,1:14], 
                                       depth = MKP.depth, 
                                       Homo=c(), 
                                       WhitelistV = MKP_WhiteList)  
    mito_master.reshuffle <- reshuffle_mito_master(mito_master)
    mito_master.list[[paste(cur_name, "All",sep='.')]] <- mito_master
    mito_master.list[[paste(cur_name, "All.reshuffle",sep='.')]] <- mito_master.reshuffle
    AOC_lis[[paste(cur_name, "All",sep='.')]] <- RunMito2Crisp_v2(whm.list[[i]], as.matrix(mito_master$d.w_jaccard), mito_master$Cts.Mtx.bi,k=10)
    phylo_lis[[paste(cur_name, "All",sep='.')]] <- mito_master$phylo
    AOC_lis[[paste(cur_name, "All.reshuffle",sep='.')]] <- RunMito2Crisp_v2(whm.list[[i]], as.matrix(mito_master.reshuffle$d.w_jaccard), mito_master.reshuffle$Cts.Mtx.bi,k=10)
    phylo_lis[[paste(cur_name, "All.reshuffle",sep='.')]] <- mito_master.reshuffle$phylo
    # count1exclude
    mito_master.count1exclude <- Make_MitoTracing_VS_count1exclude(RawGenotype.list[[i]][,1:14], 
                                                                   depth = MKP.depth, 
                                                                   Homo=c(), 
                                                                   WhitelistV = MKP_WhiteList)
    print(paste(cur_name, "count1exclude.reshuffle"))
    mito_master.count1exclude.reshuffle <- reshuffle_mito_master(mito_master.count1exclude)
    mito_master.list[[paste(cur_name, "count1exclude",sep='.')]] <- mito_master.count1exclude
    mito_master.list[[paste(cur_name, "count1exclude.reshuffle",sep='.')]] <- mito_master.count1exclude.reshuffle
    AOC_lis[[paste(cur_name, "count1exclude",sep='.')]] <- RunMito2Crisp_v2(whm.list[[i]], as.matrix(mito_master.count1exclude$d.w_jaccard), mito_master.count1exclude$Cts.Mtx.bi,k=10)
    phylo_lis[[paste(cur_name, "count1exclude",sep='.')]] <- mito_master.count1exclude$phylo
    AOC_lis[[paste(cur_name, "count1exclude.reshuffle",sep='.')]] <- RunMito2Crisp_v2(whm.list[[i]], as.matrix(mito_master.count1exclude.reshuffle$d.w_jaccard), mito_master.count1exclude.reshuffle$Cts.Mtx.bi,k=10)
    phylo_lis[[paste(cur_name, "count1exclude.reshuffle",sep='.')]] <- mito_master.count1exclude.reshuffle$phylo   

}

```
